```python
import math
import numpy as np
import matplotlib.pyplot as plt 
from scipy import signal
import pandas as pd
from scipy.fftpack import fft
```

# Trace_EVT1272070257.318228291_318.2.txt 

Señal correspondiente a los datos del anterior txt

```python
df = pd.read_csv(('/content/318.txt'), sep = " ", names = ['tiempo','tiempo_corregido','señal','b','c']) #se cargan los datos
df #lee los datos 

df.drop(['tiempo', 'b', 'c'], axis=1)
```

```python
x = df.tiempo_corregido 
y = df.señal

plt.plot(x,y, color='green'), plt.show()

plt.specgram(y,Fs=1, cmap='inferno')


plt.show()


```
